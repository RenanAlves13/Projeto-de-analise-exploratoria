```python
# Exibindo o gráfico de barras para da quantidade de alunos em cada race/ethnicity
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="ticks", font_scale=1.2)
plt.figure(figsize=(10,6))

ordem = sorted(df["race/ethnicity"].unique())

ax = sns.countplot(
    data=df,
    x="race/ethnicity",
    palette="viridis",
    order=ordem
)

plt.title("Frequência de Grupos (Raça/Etnia)")
plt.xlabel("Grupos")
plt.ylabel("Contagem Absoluta")
plt.show()
```
