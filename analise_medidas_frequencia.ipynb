```python
# Exibindo o gráfico de barras para da quantidade de alunos em cada race/ethnicity
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="ticks", font_scale=1.2)
plt.figure(figsize=(10,6))

ordem = sorted(df["race/ethnicity"].unique())

ax = sns.countplot(
    data=df,
    x="race/ethnicity",
    palette="viridis",
    order=ordem
)

plt.title("Frequência de Grupos (Raça/Etnia)")
plt.xlabel("Grupos")
plt.ylabel("Contagem Absoluta")
plt.show()
```

```python
# Exibindo na célula do jupyter os valores
df["race/ethnicity"].value_counts().sort_index()
```

```python
# Calculando a frequência relativa percentual para cada race/ethnicity
freq_relativa_percentual = df["race/ethnicity"].value_counts().sort_index().reset_index()

total = freq_relativa_percentual["count"].sum()

freq_relativa_percentual["Freq. Relativa Percentual (%)"] = (freq_relativa_percentual["count"] / total) * 100

freq_relativa_percentual.columns = ["Grupo", "Total", "Freq. Relativa Percentual (%)"]
print(freq_relativa_percentual)
```

```python
# Construindo o gráfico de barras para exibir a freq. relativa absoluta para a coluna de race/ethnicity
sns.set_theme(style="whitegrid", font_scale=0.8)

sns.barplot(
    data=freq_relativa_percentual,
    x="Grupo",
    y="Freq. Relativa Percentual (%)",
    palette="magma"
)


plt.show()
```
